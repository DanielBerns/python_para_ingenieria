```python
# Anuncio que librerias voy a usar
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
filas = 10000
columnas = 10
datos = np.random.uniform(size=(filas, columnas))
```

```python
datos.shape # forma o tamaño de los datos
```

## Que podemos hacer?

### Primera versión

```python
def explorar_dataset_v1(dataset, nombre):
    filas, columnas = dataset.shape
    print(f'Dataset {nombre:s}')
    print(f'filas {filas:d} - columnas {columnas:d}')
    min_filas = np.min(dataset, axis=1)
    max_filas = np.max(dataset, axis=1)
    mean_filas = np.mean(dataset, axis=1)
    std_filas = np.std(dataset, axis=1)
    num_filas = list()
    for v in range(filas):
        num_filas.append(v)
    num_filas = np.array(num_filas)
    plt.plot(num_filas, min_filas, '.')
    plt.plot(num_filas, max_filas, 'o')
    plt.plot(num_filas, mean_filas, 'x')
    plt.plot(num_filas, std_filas, ':')
    
    
```

```python
explorar_dataset_v1(datos, 'datos')
```

### Cuarta versión

```python
import matplotlib.gridspec as gridspec
```

```python
def explorar_dataset_v4(dataset, nombre):
    # Cálculos
    filas, columnas = dataset.shape
    print(f'Dataset {nombre:s}')
    print(f'filas {filas:d} - columnas {columnas:d}')
    min_filas = np.min(dataset, axis=1)
    max_filas = np.max(dataset, axis=1)
    mean_filas = np.mean(dataset, axis=1)
    std_filas = np.std(dataset, axis=1)
    num_filas = list()
    for v in range(filas):
        num_filas.append(v)
    num_filas = np.array(num_filas)
    
    #  Gráfico
    fig = plt.figure(figsize=(7, 11))
    gs = gridspec.GridSpec(nrows=2, ncols=2, height_ratios=[2, 2])

    #  min_filas
    ax0 = fig.add_subplot(gs[0, 0])
    ax0.plot(num_filas, min_filas, '.')
    ax0.set_title('min_filas')

    # max_filas
    ax1 = fig.add_subplot(gs[0, 1])
    ax1.plot(num_filas, max_filas, '.')
    ax1.set_title('max_filas')

    #  mean_filas
    ax2 = fig.add_subplot(gs[1, 0])
    ax2.plot(num_filas, mean_filas, '.')
    ax2.set_title('mean_filas')

    # std_filas
    ax3 = fig.add_subplot(gs[1, 1])
    ax3.plot(num_filas, std_filas, '.')
    ax3.set_title('std_filas')

    plt.tight_layout()
    plt.show()
```

```python
explorar_dataset_v4(datos, 'datos')
```
