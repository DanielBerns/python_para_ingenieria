```python
# Anuncio que librerias voy a usar
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline # para que no haya que agregar plt.show()
```

```python
filas = 10000
columnas = 10
datos = np.random.uniform(size=(filas, columnas))
```

```python
datos.shape # forma o tamaño de los datos
```

```python
datos.shape == (filas, columnas)
```

## Que podemos hacer?

```python
# valor minimo de todos los números en datos
min_total = np.min(datos)
min_total
```

```python
min_total.shape
```

```python
# valor máximo de todos los números en datos
max_total = np.max(datos)
max_total
```

```python
max_total.shape
```

```python
# promedio de todos los numeros en datos
promedio_total = np.mean(datos)
promedio_total.shape
```

```python
# promedio por filas
promedio_por_filas = np.mean(datos, axis=1)
promedio_por_filas.shape == (filas,)  # Verifico que la forma del resultado sea la esperada
```

```python
# promedio por columnas
promedio_por_columnas = np.mean(datos, axis=0)
assert(promedio_por_columnas.shape == (columnas,))
```

```python
# desviacion estandar de todos los numeros en datos
desviacion_estandar_total = np.std(datos)
desviacion_estandar_total.shape == ()
```

```python
# desviacion estandar por filas
desviacion_estandar_por_filas = 0                 # reemplazar
assert(desviacion_estandar_por_filas.shape == (filas,))
```

```python
# desviacion estandar por columnas
desviacion_estandar_por_columnas = 0                 # reemplazar
assert(desviacion_estandar_por_columnas.shape == (columnas,))
```
