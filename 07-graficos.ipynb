```python
# Anuncio que librerias voy a usar
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
filas = 1000
columnas = 1
datos = np.random.uniform(size=(filas, columnas))
n = np.arange(filas)
```

```python
plt.plot(n, datos[:,0],'.')
plt.plot(n, np.mean(datos[:,0]) * np.ones(n.shape),'r.')
acumulado = np.cumsum(datos[:,0]) / (n + 1)
plt.plot(n, acumulado, '.')
```

```python
def demo_convergencia_promedio(cantidad):
    datos = np.random.uniform(size=cantidad)
    n = np.arange(cantidad)
    promedio_real = np.mean(datos) * np.ones(n.shape)
    aproximado = np.cumsum(datos) / (n + 1)
    plt.plot(n, datos,'.')
    plt.plot(n, promedio_real,'r.')
    plt.plot(n, aproximado, '.')
```

```python
f = 1
for k in range(50, 1000, 50):
    plt.figure(f)
    demo_convergencia_promedio(k)
    f += 1
```

```python
def demo_convergencia_promedio_v1(datos):
    n = np.arange(datos.shape[0])
    promedio_real = np.mean(datos) * np.ones(n.shape)
    aproximado = np.cumsum(datos) / (n + 1)
    plt.plot(n, datos,'.')
    plt.plot(n, promedio_real,'r.')
    plt.plot(n, aproximado, '.')
```

```python
datos_normal = np.random.normal(size=100)
plt.figure(1)
demo_convergencia_promedio_v1(datos_normal)
datos_uniforme = np.random.uniform(size=100)
plt.figure(2)
demo_convergencia_promedio_v1(datos_uniforme)
```
