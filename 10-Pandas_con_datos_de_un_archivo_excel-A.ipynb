```python
df = archivo_excel[['dia_j', 'Tsoil1']]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.clf() # clf abrevia clear figure
plt.plot(df.Tsoil1, '.')
plt.title('Tsoil1')
plt.xlabel('tiempo')
plt.ylabel('Temperatura')
plt.show()
```

```python
plt.clf() # clf abrevia clear figure
plt.plot(df.Tsoil1[df.Tsoil1 > 0], '.')
plt.title('Tsoil1')
plt.xlabel('tiempo')
plt.ylabel('Temperatura')
plt.show()
```

```python
plt.clf() # clf abrevia clear figure
plt.plot(df.Tsoil1[(df.Tsoil1 > 0) & (df.Tsoil1 < 25)], '.')
plt.title('Tsoil1')
plt.xlabel('tiempo')
plt.ylabel('Temperatura')
plt.show()
```
